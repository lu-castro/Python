<h3>Comprenhensions</h3>
<h8>En python las listas con colecciones de datos encerradas en paréntesis, separados por comentarios. Las List Comprenhension son listas que no encierran datos, sino que contienen bucles y clases if </h8>

    [expr for val in collection] 
    #forma más simple de List Comprenhension 

<h8>expr genera los elementos en la lista. Lo demás genera un bucle con una colección de data</h8>

    [expr for val in collection if <test>]
    [expr for val in collection if <test1> and <test2>] 
    #se ejecutará la lista si y sólo si se cumple la condición 

<h5>Lista al cuadrados</h5>

```python
square = []
for i in range(1, 101):
    square.append(i**2) #append() devuelve un ítem de la lista existente

    print(square)
```

```python
square2 = [i**2 for i in range(1, 101)]
print(square2)
```

```python
remainders5 = [x**2 % 5 for x in range(1, 101)]
print(remainders5)
```

```python
remainders11 = [x**2 % 11 for x in range(1,101)]
print(remainders11)
```
